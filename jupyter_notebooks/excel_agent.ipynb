````python
#  @title  define the agent
from smolagents import CodeAgent, OpenAIServerModel
from google.colab import userdata

model = OpenAIServerModel(model_id ="gemini-2.0-flash",
                          api_base = "https://generativelanguage.googleapis.com/v1beta/openai/",
                          api_key = userdata.get('GOOGLE_API_KEY'), # my api key, keep as secret
                          )
agent = CodeAgent(tools=[], model=model,
                  additional_authorized_imports=["numpy","pandas"]
                  )
````

````python
#  @title run the agent with human instruction, change the prompt to test it
prompt='''
read /content/sample_data/california_housing_test.csv,
sort the table by median_income in descending order,
and save the first ten rows to  a new file named output.csv
'''
agent.run(prompt)
````
